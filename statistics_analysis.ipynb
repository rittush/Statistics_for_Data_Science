```python

import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm

sales_data = pd.read_csv(r"C:\Users\Rittush\Downloads\sales_data.csv")
customer_churn = pd.read_csv(r"C:\Users\Rittush\Downloads\customer_churn.csv")

sales_data_head = sales_data.head()
customer_churn_head = customer_churn.head()

sales_desc = sales_data.describe(include='all')
sales_data_numeric = sales_data.select_dtypes(include=[np.number])
sales_skew = sales_data_numeric.skew()
sales_kurtosis = sales_data_numeric.kurt()

correlation_matrix = sales_data_numeric.corr()
if 'Sales' in sales_data.columns:
    ttest_sales = stats.ttest_1samp(sales_data['Total_Sales'], 0)
else:
    ttest_sales = "Sales column not found"

if 'Marketing_Spend' in sales_data.columns and 'Revenue' in sales_data.columns:
    corr_test = stats.pearsonr(sales_data['Marketing_Spend'], sales_data['Revenue'])
else:
    corr_test = "Required columns not found"

if 'Sales' in sales_data.columns and 'Marketing_Spend' in sales_data.columns:
    median_spend = sales_data['Marketing_Spend'].median()
    high_spend = sales_data[sales_data['Marketing_Spend'] > median_spend]['Total_Sales']
    low_spend = sales_data[sales_data['Marketing_Spend'] <= median_spend]['Total_Sales']
    ttest_independent = stats.ttest_ind(high_spend, low_spend, equal_var=False)
else:
    ttest_independent = "Required columns not found"

confidence_intervals = {}
for col in sales_data_numeric.columns:
    mean = sales_data_numeric[col].mean()
    sem = stats.sem(sales_data_numeric[col])
    ci = stats.t.interval(0.95, len(sales_data_numeric[col])-1, loc=mean, scale=sem)
    confidence_intervals[col] = ci

if 'Marketing_Spend' in sales_data.columns and 'Revenue' in sales_data.columns:
    X = sales_data['Marketing_Spend']
    y = sales_data['Revenue']
    X = sm.add_constant(X)
    model = sm.OLS(y, X).fit()
    regression_summary = model.summary()
else:
    regression_summary = "Required columns not found"

sales_data_head, customer_churn_head, sales_desc, sales_skew, sales_kurtosis, correlation_matrix, ttest_sales, corr_test, ttest_independent, confidence_intervals, regression_summary
```
